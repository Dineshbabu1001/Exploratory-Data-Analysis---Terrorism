```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.express as px
import plotly.colors as colors
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("/content/global_terrorism.csv", encoding='latin1')
```

```python
data.head()
```

```python
data.columns.values
```

```python
data.rename(columns={'iyear':'Year',
                     'imonth':'Month',
                     'iday':"day",
                     'gname':'Group',
                     'country_txt':'Country',
                     'region_txt':'Region',
                     'provstate':'State',
                     'city':'City',
                     'latitude':'latitude',
                     'longitude':'longitude',
                     'summary':'summary',
                     'attacktype1_txt':'Attacktype',
                     'targtype1_txt':'Targettype',
                     'weaptype1_txt':'Weapon',
                     'nkill':'kill',
                     'nwound':'Wound'},inplace=True)
```

```python
data = data[['Year','Month',
             'day','Country',
             'State','Region',
             'City','latitude',
             'longitude',"Attacktype",
             'kill','Wound',
             'target1','summary',
             'Group','Targettype',
             'Weapon','motive']]
```

```python
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data['Wound'] = data['Wound'].fillna(0)
data['kill'] = data['kill'].fillna(0)
```

```python
data['Casualities'] = data['kill'] + data['Wound']
```

```python
data.info()
```

```python
data.describe()
```

```python
years_count = data['Year'].value_counts().sort_index()
```

```python
fig = px.bar(x=years_count.index,
             y=years_count.values,
             labels={'x': 'Year', 'y': 'Number of Attacks'},
             title='Number of Terrorist Attacks Each Year')

fig.update_layout(xaxis=dict(tickangle=45))
fig.show()
```

```python
cross_tab = pd.crosstab(data['Year'], data['Region'])
```

```python
cross_tab = cross_tab.reset_index()
```

```python
fig = px.area(cross_tab,
              x='Year',
              y=cross_tab.columns[1:],
              title='Terrorist Activities By Region In Each Year',
              labels={'Year': 'Year', 'value': 'Number of Attacks', 'variable': 'Region'})
```

```python
fig.update_layout(
    xaxis=dict(title=dict(text='Year', font=dict(size=20))),
    yaxis=dict(title=dict(text='Number of Attacks', font=dict(size=20))),
    font=dict(size=20))
fig.show()
```

```python
data.Group.value_counts()[1:10]
```

```python
fig = px.bar(data_frame=data['Country'].value_counts()[:10],
             x=data['Country'].value_counts()[:10].index,
             y=data['Country'].value_counts()[:10].values,
             color=data['Country'].value_counts()[:10].index,
             color_discrete_sequence=px.colors.sequential.YlOrBr_r,
             labels={'x': 'Countries', 'y': 'Count'},
             title='Top Countries Affected',
             height=500,
             width=800)
fig.update_xaxes(tickangle=45)
fig.show()
```

```python
df = data[['Year', 'kill']].groupby(['Year']).sum().reset_index()

fig = px.bar(df,
             x='Year',
             y='kill',
             labels={'Year': 'Year', 'kill': 'Number of killed people'},
             title="People Died Due To Attack",
             height=500, width=800)

fig.update_xaxes(tickangle=45,
                 showline=True,
                 linewidth=1,
                 linecolor='black')
fig.update_yaxes(showline=True,
                 linewidth=1,
                 linecolor='black')
fig.update_layout(xaxis_title="Year",
                  yaxis_title="Number of killed people")

fig.show()
```

```python
city_counts = data['City'].value_counts().to_frame().sort_values('City', axis=0, ascending=False).head(10)

fig = px.bar(city_counts,
             x=city_counts.index,
             y='City',
             labels={'x': 'City', 'City': 'Number of attacks'},
             color='City',
             color_continuous_scale='Blues',
             title='Top 10 most affected cities',
             height=500, width=800)

fig.update_xaxes(tickangle=45)
fig.update_layout(xaxis_title="City", yaxis_title="Number of attacks")

fig.show()
```

```python
attacktype_counts = data['Attacktype'].value_counts()

fig = px.bar(attacktype_counts,
             x=attacktype_counts.index,
             y=attacktype_counts.values,
             labels={'x': 'Attack Type', 'y': 'Number of Attacks'},
             color=attacktype_counts.index,
             color_discrete_sequence=px.colors.qualitative.Dark2,
             title='Name of Attack Types',
             height=500, width=800)

fig.update_xaxes(tickangle=45)
fig.update_layout(xaxis_title="Attack Type", yaxis_title="Number of Attacks")

fig.show()
```

```python
kill_by_attacktype = data[['Attacktype', 'kill']].groupby(["Attacktype"], axis=0).sum()

fig = px.bar(kill_by_attacktype,
             x=kill_by_attacktype.index,
             y='kill',
             labels={'x': 'Attack Type', 'kill': 'Number of People Killed'},
             color=kill_by_attacktype.index,
             title='Number of People Killed by Attack Type',
             height=500, width=800)

fig.update_xaxes(tickangle=45)
fig.update_layout(xaxis_title="Attack Type", yaxis_title="Number of People Killed")

fig.show()
```

```python
wound_by_attacktype = data[['Attacktype', 'Wound']].groupby(["Attacktype"], axis=0).sum()

fig = px.bar(wound_by_attacktype,
             x=wound_by_attacktype.index,
             y='Wound',
             labels={'x': 'Attack Type', 'Wound': 'Number of People Wounded'},
             color=wound_by_attacktype.index,
             title='Number of People Wounded by Attack Type',
             height=500, width=800)

fig.update_xaxes(tickangle=45)
fig.update_layout(xaxis_title="Attack Type", yaxis_title="Number of People Wounded")

fig.show()
```

```python
targettype_counts = data['Targettype'].value_counts()

fig = px.bar(targettype_counts,
             x=targettype_counts.index,
             y=targettype_counts.values,
             labels={'x': 'Target Type', 'y': 'Count'},
             color=targettype_counts.index,
             title='Number of Attacks by Target Type',
             height=500, width=800)

fig.update_xaxes(tickangle=90)
fig.update_layout(xaxis_title="Target Type", yaxis_title="Count")

fig.show()
```

```python
group_counts = data['Group'].value_counts().to_frame().drop('Unknown').head(10)

fig = px.bar(group_counts,
             x=group_counts.index,
             y='Group',
             labels={'x': 'Terrorist Group Name', 'Group': 'Attack Number'},
             color=group_counts.index,
             title='Top 10 Terrorist Group Attacks',
             height=500, width=800)

fig.update_xaxes(tickangle=45)
fig.update_layout(xaxis_title="Terrorist Group Name",
                  yaxis_title="Attack Number")

fig.show()
```

```python
group_kill = data[['Group', 'kill']].groupby(['Group'], axis=0).sum().drop('Unknown').sort_values('kill', ascending=False).head(10)

fig = px.bar(group_kill,
             x=group_kill.index,
             y='kill',
             labels={'x': 'Terrorist Group Name', 'kill': 'Number of Killed People'},
             color=group_kill.index,
             title='Top 10 Terrorist Group Attacks (by Number of Killed People)',
             height=500, width=800)

fig.update_xaxes(tickangle=45)
fig.update_layout(xaxis_title="Terrorist Group Name",
                  yaxis_title="Number of Killed People")

fig.show()
```
